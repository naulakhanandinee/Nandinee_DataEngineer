Defining headers for csv file


```python
csv_headers = [
    'FinInstrmGnlAttrbts.Id',
    'FinInstrmGnlAttrbts.FullNm',
    'FinInstrmGnlAttrbts.ClssfctnTp',
    'FinInstrmGnlAttrbts.CmmdtyDerivInd',
    'FinInstrmGnlAttrbts.NtnlCcy',
    'Issr'
]
```

Parsing the XML File

```python
tree=ET.parse('//content/drive/MyDrive/Colab Notebooks/data1.xml')
root=tree.getroot()
```

Opening the csv file to write


```python
with open('output.csv', 'w', newline='') as csvfile:
    writer = csv.writer(csvfile)
    writer.writerow(csv_headers)
```

Extract data from XML file


```python
 for fin_instrm in root.findall('.//FinInstrmGnlAttrbts'):
       
        id = fin_instrm.find('Id').text
        full_nm = fin_instrm.find('FullNm').text
        clssfctn_tp = fin_instrm.find('ClssfctnTp').text
        cmmdty_deriv_ind = fin_instrm.find('CmmdtyDerivInd').text
        ntnl_ccy = fin_instrm.find('NtnlCcy').text
        issr = fin_instrm.find('../Issr').text

        writer.writerow([id, full_nm, clssfctn_tp, cmmdty_deriv_ind, ntnl_ccy, issr])
```
